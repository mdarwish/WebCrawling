```python
def extract_fight_metrics(driver):
    # The fight card loop. This loop will execute for every event page
    matches = driver.find_elements_by_class_name("fight")
    page_soup = soup(driver.page_source, "html.parser")
    event_card = page_soup.find("div", {"id":"fightCard"})
    card_actions = []
    card_images = []
    for match in matches:
        try:
            match.click()
            sleep(5)
        except ElementNotVisibleException:
            continue
        fight = soup(match.get_attribute('innerHTML'), "html.parser")
        page_soup = soup(driver.page_source, "html.parser")
        try:
            fightStatDetails = (page_soup.find("div", {"id":"fightStatDetailsInner"}))
        except:
            print('"id":"fightStatDetailsInner" not found.')
            fightStatDetails = (page_soup.find("div", {"id":"fightDetailsInner"}))
            print(fightStatDetails)
        fighterInfo = (page_soup.find("div", {"id":"fighterInfo"}))
        action_record = []
        fight_images_sml = []
        fight_imgs = (fight.find_all('div', {"class":"fighter-image-frame"}))
        right_img, left_img = (fight_imgs[0]).find('img'), (fight_imgs[1]).find('img')
        # print(right_img, left_img)
        right_img_url, left_img_url = 'https://' + (right_img.attrs['src'])[2:], 'https://' + (left_img.attrs['src'])[2:]
        right_img_filename, left_img_filename = (right_img_url.split('/'))[-1], (left_img_url.split('/'))[-1]
        fight_images_sml = [{'right_img_filename': right_img_filename, 'right_img_url':right_img_url}, 
                            {'left_img_filename': left_img_filename, 'left_img_url':left_img_url}]
        fight_images_standing = []
        if fightStatDetails != None:
            standing_images = fightStatDetails.find_all("div", {"class":"fighter-image"})
            fighter1_image = 'https://' + ((standing_images[0]).find("img").attrs['src'])[2:]
            fighter2_image = 'https://' + ((standing_images[1]).find("img").attrs['src'])[2:]
            fighter1_image_file, fighter2_image_file = (fighter1_image.split('/'))[-1], (fighter2_image.split('/'))[-1]
            fight_images_standing = [{'fighter1_image_file': fighter1_image_file, 'fighter1_image':fighter1_image}, 
                                     {'fighter2_image_file': fighter2_image_file, 'fighter2_image':fighter2_image}]
        card_images.append([fight_images_sml, fight_images_standing])
        # print(card_images)
        try:
            fightStats = (page_soup.find("table", {"id":"fightStats"}))
            if fightStats != None:
                fighter1_Control = (fightStats.find("tr", {"id":"fighterControl"})).find("td", {"class": "left"}).get_text()
                fighter2_Control = (fightStats.find("tr", {"id":"fighterControl"})).find("td", {"class": "right"}).get_text()
                fighter1_Knockdowns = (fightStats.find("tr", {"id":"fighterKnockdowns"})).find("td", {"class": "left"}).get_text()
                fighter2_Knockdowns = (fightStats.find("tr", {"id":"fighterKnockdowns"})).find("td", {"class": "right"}).get_text()
                fighter1_totalStrikes = re.sub('\n',' ',(fightStats.find("tr", {"id":"totalStrikes"})).find("td", {"class": "left"}).get_text())
                fighter2_totalStrikes = re.sub('\n',' ',(fightStats.find("tr", {"id":"totalStrikes"})).find("td", {"class": "right"}).get_text())
                fighter1_significantStrikes = re.sub('\n',' ',(fightStats.find("tr", {"id":"significantStrikes"})).find("td", {"class": "left"}).get_text())
                fighter2_significantStrikes = re.sub('\n',' ',(fightStats.find("tr", {"id":"significantStrikes"})).find("td", {"class": "right"}).get_text())
                fighter1_takedowns = re.sub('\n',' ',(fightStats.find("tr", {"id":"takedowns"})).find("td", {"class": "left"}).get_text())
                fighter2_takedowns = re.sub('\n',' ',(fightStats.find("tr", {"id":"takedowns"})).find("td", {"class": "right"}).get_text())    
                fighter1_subAttempts = re.sub('\n',' ',(fightStats.find("tr", {"id":"subAttempts"})).find("td", {"class": "left"}).get_text())
                fighter2_subAttempts = re.sub('\n',' ',(fightStats.find("tr", {"id":"subAttempts"})).find("td", {"class": "right"}).get_text())
                first_names = fighterInfo.find_all("div", {'class': 'fighter-first-name'})
                last_names = fighterInfo.find_all("div", {'class': 'fighter-last-name'})
                fighter1_name = (first_names[0]).get_text() + " " + (last_names[0]).get_text()
                fighter2_name = (first_names[1]).get_text() + " " + (last_names[1]).get_text()

                action_record = [fighter1_name, fighter1_Control, fighter1_Knockdowns, 
                                 fighter1_totalStrikes, fighter1_significantStrikes, 
                                 fighter1_takedowns, fighter1_subAttempts, 
                                 fighter2_name, fighter2_Control, fighter2_Knockdowns, 
                                 fighter2_totalStrikes, fighter2_significantStrikes, 
                                 fighter2_takedowns, fighter2_subAttempts]
            else:
                action_record = [''] * 14

            card_actions.append(action_record)
        except:
            continue

        
        # print(action_record)
    return card_actions, card_images
```

```python
def downloader(image_url, f_name=None):
    file_name = random.randrange(1,10000)
    if f_name == None:
        full_file_name = str(file_name) + '.jpg'
    else:
        full_file_name = f_name

    try:
        r = requests.get(image_url)

        with open(full_file_name, 'wb') as f:
            f.write(r.content)
    except:
        return False
    
    return True
```
